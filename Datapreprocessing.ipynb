```python
def minmaxNormalize(data):
  norm_target= np.empty((32, 32, 16,2), dtype=np.float32)
  for l in range(1, data.shape[3]):
    for i in range(1, data.shape[2]):
      temp=(data[:,:,i,l]-np.min(data[:,:,i,l]))/(np.max(data[:,:,i,l])-np.min(data[:,:,i,l]))
      norm_target[:,:,i,l]=temp
  return norm_target
```

```python
def roi_weighted_matrix(roi,roifactor):
  roi_wt_mask_ard=np.array(roi, dtype=float)
  roi_wt_mask_ard[roi_wt_mask_ard==0]=roifactor
  roi_wt_mask_ard[roi_wt_mask_ard==1]=(1-roifactor)
  return roi_wt_mask_ard
```

```python
def downsample(input_array,downsample_factor):
  from scipy.interpolate import RegularGridInterpolator
  steps = [1/downsample_factor, 1/downsample_factor, 1/downsample_factor]    # original step sizes
  x, y, z = [steps[k] * np.arange(input_array.shape[k]) for k in range(3)]  # original grid
  f = RegularGridInterpolator((x, y, z), input_array)    # interpolator
  f
  dx, dy, dz = 1.0, 1.0, 1.0    # new step sizes
  new_grid = np.mgrid[0:x[-1]:dx, 0:y[-1]:dy, 0:z[-1]:dz]   # new grid
  new_grid = np.moveaxis(new_grid, (0, 1, 2, 3), (3, 0, 1, 2))  # reorder axes for evaluation
  dwnsmpld_values = f(new_grid)
  return dwnsmpld_values
```

```python
def data_preprocess(amp,phase,target,roi,downsample_factor):
  norm_target=minmaxNormalize(target)
  this_log_amplitude=np.log(amp)
  z_amp_noisy=(this_log_amplitude-mean_amp)/std_amp
  z_phase_noisy=(phase-mean_phase)/std_phase
  roi_factor=(roi==1).sum()/(roi==0).sum()
  roi_weighted_mask=roi_weighted_matrix(roi,roi_factor)
  dwnsmpl_target=downsample(norm_target,downsample_factor)
  dwnsmpl_wt_roi=downsample(roi_weighted_mask,downsample_factor)
  return this_log_amplitude,norm_target,z_amp_noisy,z_phase_noisy,roi_factor,roi_weighted_mask,dwnsmpl_target,dwnsmpl_wt_roi
```

```python
for i in range(1,nfiles+1):
  example=globals()["dict_"+str(i)]
  globals()["processed_data_"+str(i)]=data_preprocess(example['amplitude_noisy'],example['phase_noisy'],example['target'],example['roi_mask'],2)
```
